```python
import pandas as pd
```

```python
df = pd.read_csv('salaries_by_college_major.csv')
```

```python
df.head()
```

```python
# how many rows and columns
df.shape
```

```python
# What are the labels for the columns? Do the columns have names?
df.columns
```

```python
# find nulls
df.isna()
clean_df = df.dropna()
```

```python
# find the college major with the highest mid-career salary
clean_df[['Mid-Career Median Salary','Undergraduate Major']].sort_values(by='Mid-Career Median Salary' ,ascending=False)
```

```python
# find lowest starting salary degree
clean_df.sort_values(by='Starting Median Salary',ascending=True)
```

```python
# Which college major has the lowest mid-career salary and how much can people expect to earn with this degree? 

clean_df.sort_values(by='Mid-Career Median Salary',ascending=True)
```

```python
clean_df['Low Risk'] = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
#clean_df['Low Risk'] = 
low_risk = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
clean_df.insert(1,'Riskiness',low_risk)
```

```python
clean_df.head()
```

```python
# highest earning potential
clean_df.sort_values(by='Mid-Career 90th Percentile Salary',ascending=False)
```

```python
# highest spread
clean_df.sort_values(by='Riskiness',ascending=False)
```

```python
# grouping practice
clean_df.groupby('Group').mean()
```

```python
# setting for numbers
pd.options.display.float_format = '{:,.0f}'.format
```

```python
# stretch goal - make a scatterplot of 
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10
data = clean_df
labels =  data['Undergraduate Major']#['point{0}'.format(i) for i in range(N)]
plt.figure(figsize=(20,20))
plt.subplots_adjust(bottom = 0.1)
plt.scatter(
    data['Starting Median Salary'], data['Mid-Career 90th Percentile Salary'], marker='o',
    cmap=plt.get_cmap('Spectral'))

for label, x, y in zip(labels, data['Starting Median Salary'], data['Mid-Career 90th Percentile Salary']):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-20, 20),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
plt.xlabel('Starting Median Salary')
plt.ylabel('Mid-Career 90th Percentile Salary')

plt.title('Major: Entry Level vs. Upside')


plt.show()
```
