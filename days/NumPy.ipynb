### Broadbasting is when you use a scalar (read: constant) and then scale it up to the size of whatever array youre using

```python
a10 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])
 
b10 = np.array([[4, 1, 3],
               [5, 8, 5]])
```

```python
result = np.matmul(a10,b10)
```
